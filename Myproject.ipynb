```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dataset= pd.read_csv('red_wine.csv')
```

```python
columns_to_plot =['citric acid','quality']
plt.figure (figsize=(10,6))
for column in columns_to_plot:
    plt.hist(dataset[column],bins=30,alpha=0.5,label=column)
    plt.title('citric acid vs quality')
    plt.xlabel ('citric acid')
    plt.ylabel ('quality')
    plt.legend (loc='upper right')
    plt.show()
```

```python
columns_to_plot=['alcohol','quality']
plt.figure(figsize=(10,60))
for column in columns_to_plot:
    plt.hist(dataset[column],bins=30,alpha=0.5,label=column)
    plt.title('A graph of alcohol against quality')
    plt.xlabel('alcohol')
    plt.ylabel('quality')
    plt.legend(loc='upper right')
    plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(dataset['density'],dataset['fixed acidity'],color='blue',alpha=0.5)
plt.title('density vs.fixed acidity')
plt.xlabel('Density')
plt.ylabel('Fixed Acidity')
plt.grid(True)
plt.show()
```

```python
avg_pH_by_quality = dataset.groupby('quality')['pH'].mean()
plt.figure(figsize=(8, 6))
avg_pH_by_quality.plot(kind='bar', color='brown')
plt.title('Average pH by Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Average pH')
plt.grid(axis='y')
plt.show()
```

### **A REPORT ON GRAPHS AND PLOTS OBTAINED**
---
#### **A Graph(Histogram) of Citric acid vs Quality**
##### The histogram displays the distribution of wine quality rating in the red wine based on the citric acid content of the red wine dataset.The x-axis represents the range for citric acid content, while the y-axis indicates the frequency of wines falling within each citric acid content range for various quality levels.The histogram helps us observe how citric acid content is distributed among different quality ratings, and also whether or not quality ratings tend to have higher or lower citric acid content.

#### **A Graph(Histogram) of Alcohol against Quality**
##### The histogram displays the distribution of the quality of red wine based on its alcohol content.The x-axis represents range of alcohol percentage present in the dataset and the y-axis represents the number of occurrences for each range of alcohol content within specific quality ratings.The histogram shows how certain quality levels are concentrated within specific alcohol ranges.

#### **A bar chart of pH vs. quality**
##### The bar chart plotted illustrates the pH level of each wine quality rating present in the red wine dataset. Each bar represents the mean pH value corresponding to a specific wine quality level. The x-axis represents the quality and the y-axis represents the average pH of red wine.The title of this graph is Average pH by Wine Quality.


#### **A scatter Diagram/Plot of Density vs. Fixed Acidity**
##### The scatter plot illustrates the relationship between the density and fixed acidity of wines in the dataset.The x-axis represents the fixed acidity values and the y-axis represents the densityy values.The scatter plot shows a strong correlation between the two variables because of the noticeable trend where an increase or decrease in fixed acidity corresponds to a similar trend in density.
